### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot
# https://matplotlib.org/stable/users/explain/quick_start.html
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
# fig, ax = plt.subplots()
plt.title('Sample Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20,'color':'blue'})
plt.xlabel('X axis')
plt.ylabel('Y axis')
# ax.set_title("Sample Graph")
# ax.set_xlabel('x label')
# ax.set_ylabel('y label')
plt.plot(x,y)
plt.show() # to remove that line comes with output ([<matplotlib.lines.Line2D at 0x28ece4996d8>])
```

```python
# change tick marks, add legend
x = [0,1,2,3,4]
y = [0,2,4,6,8]
# fig, ax = plt.subplots()
plt.title('Sample Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20,'color':'blue'})
plt.xlabel('X axis')
plt.ylabel('Y axis')
# ax.set_title("Sample Graph")
# ax.set_xlabel('x label')
# ax.set_ylabel('y label')

plt.xticks([0,1,2,3])  # change tick marks
plt.yticks([0,2,4,6,8,10])

# plt.plot(x,y, label='2x',color='red',linewidth=3,marker='.',markersize=10,markeredgecolor='blue',linestyle="--") # label name
# use shorthand notation for this
# [color,marker,line]
plt.plot(x,y, 'r.-',label='2x')

plt.legend() # add legend

plt.show() # to remove that line comes with output ([<matplotlib.lines.Line2D at 0x28ece4996d8>])
```

```python
# add a second line to the graph
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# resize the graph
plt.figure(figsize=(5,3),dpi=300) # dpi pixels per inch, recommended 300

plt.title('Sample Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20,'color':'blue'})
plt.xlabel('X axis')
plt.ylabel('Y axis')

x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,label='x^2')
# print(x2)

plt.xticks([0,1,2,3])  
plt.yticks([0,2,4,6,8,10])

plt.plot(x,y, 'r.-',label='2x')

plt.legend()
plt.savefig('my_graph.png',dpi=300) # save the graph
plt.show() 
```

### Bar chart

```python
labels = ['A','B','C']
values = [3,2,7]
bars = plt.bar(labels,values)

# 1 way
# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

# 2nd way
# patterns = ['/','O','*']
# for bar in bars:
#     bar.set_hatch(patterns.pop(0))

# 3rd way
patterns = ['/','O','*']
for index,pattern in enumerate(patterns):
    bars[index].set_hatch(pattern)

plt.figure(figsize=(6,4))
plt.show()
```

### Real World Examples

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('data_files/gas_prices.csv')
```

```python
gas.head()
```

```python
plt.figure(figsize=(8,5)) # put this in top

plt.title('Gas prices over time (in USD)')

# plt.plot(gas.Year, gas.USA,'b.-')
# plt.plot(gas.Year, gas.Australia,'y.-')
# plt.plot(gas['Year'], gas.Canada,'r.-', label='United States')
# plt.plot(gas['Year'], gas['South Korea'], 'g.-')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.')


plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()
plt.show()
```

```python
# place the legend outside matplotlib graph
# Create a matplotlib figure
fig, ax = plt.subplots()


plt.title('Gas prices over time (in USD)')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.')


plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')

# Add a legend
pos = ax.get_position()
ax.set_position([pos.x0, pos.y0, pos.width * 0.9, pos.height])
ax.legend(loc='center right', bbox_to_anchor=(1.35, 0.5))

plt.show()
```
