### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
# fig, ax = plt.subplots()
plt.title('Sample Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20,'color':'blue'})
plt.xlabel('X axis')
plt.ylabel('Y axis')
# ax.set_title("Sample Graph")
# ax.set_xlabel('x label')
# ax.set_ylabel('y label')
plt.plot(x,y)
plt.show() # to remove that line comes with output ([<matplotlib.lines.Line2D at 0x28ece4996d8>])
```

```python
# change tick marks, add legend
x = [0,1,2,3,4]
y = [0,2,4,6,8]
# fig, ax = plt.subplots()
plt.title('Sample Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20,'color':'blue'})
plt.xlabel('X axis')
plt.ylabel('Y axis')
# ax.set_title("Sample Graph")
# ax.set_xlabel('x label')
# ax.set_ylabel('y label')

plt.xticks([0,1,2,3])  # change tick marks
plt.yticks([0,2,4,6,8,10])

# plt.plot(x,y, label='2x',color='red',linewidth=3,marker='.',markersize=10,markeredgecolor='blue',linestyle="--") # label name
# use shorthand notation for this
# [color,marker,line]
plt.plot(x,y, 'r.-',label='2x')

plt.legend() # add legend

plt.show() # to remove that line comes with output ([<matplotlib.lines.Line2D at 0x28ece4996d8>])
```

```python
# add a second line to the graph
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# resize the graph
plt.figure(figsize=(5,3),dpi=300) # dpi pixels per inch, recommended 300

plt.title('Sample Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20,'color':'blue'})
plt.xlabel('X axis')
plt.ylabel('Y axis')

x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,label='x^2')
# print(x2)

plt.xticks([0,1,2,3])  
plt.yticks([0,2,4,6,8,10])

plt.plot(x,y, 'r.-',label='2x')

plt.legend()
plt.savefig('my_graph.png',dpi=300) # save the graph
plt.show() 
```

```python
# place the legend outside matplotlib graph
# Create a matplotlib figure
fig, ax = plt.subplots()


plt.title('Gas prices over time (in USD)')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.')


plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')

# Add a legend
pos = ax.get_position()
ax.set_position([pos.x0, pos.y0, pos.width * 0.9, pos.height])
ax.legend(loc='center right', bbox_to_anchor=(1.35, 0.5))

plt.show()
```

```python
fifa.Weight = [int(str(x).strip('lbs')) if type(x) == str else x for x in fifa.Weight]
```

```python
fifa.Weight
```

```python
light = fifa[fifa.Weight < 125].shape[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].shape[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].shape[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].shape[0]
heavy = fifa[fifa.Weight > 200].shape[0]
```

```python
medium_heavy
```

```python
plt.style.use('ggplot')
weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125','125 - 150','150 - 175','175 - 200','Over 200']
explode = (.1,.1,0,0,.4)
plt.title("Weight distribution of fifa player(in ibs)")
plt.pie(weights, labels = labels,autopct="%.2f %%",pctdistance=0.8,explode=explode)
plt.show()
```

```python
plt.style.use('default')
labels = ['FC Barcelona','Real Madrid']
plt.title('Professional Football Team Comparison')
boxes = plt.boxplot([barcelona, madrid], labels=labels, patch_artist=True,medianprops={"linewidth":2})
for box in boxes['boxes']:
    box.set(color='#ffabbd',linewidth=4) # set edge color
    box.set(facecolor='#524234')
plt.show()
```

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
xs = [i for i, _ in enumerate(movies)]
# plot bars with left x-coordinates [xs], heights [num_oscars]
plt.bar(xs, num_oscars)
plt.ylabel("# of Academy Awards")
plt.title("My Favorite Movies")
# label x-axis with movie names at bar centers
plt.xticks([i for i, _ in enumerate(movies)], movies)
plt.show()
```

```python
from matplotlib import pyplot as plt
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]
# we can make multiple calls to plt.plot
# to show multiple series on the same chart
plt.plot(xs, variance,
'g-', label='variance')
# green solid line
plt.plot(xs, bias_squared, 'r-.', label='bias^2')
# red dot-dashed line
plt.plot(xs, total_error, 'b:', label='total error') # blue dotted line
# because we've assigned labels to each series
# we can get a legend for free
# loc=9 means "top center"
plt.legend(loc=9)
plt.xlabel("model complexity")
plt.title("The Bias-Variance Tradeoff")
plt.show()
```
