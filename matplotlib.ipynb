### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot
# https://matplotlib.org/stable/users/explain/quick_start.html
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
# fig, ax = plt.subplots()
plt.title('Sample Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20,'color':'blue'})
plt.xlabel('X axis')
plt.ylabel('Y axis')
# ax.set_title("Sample Graph")
# ax.set_xlabel('x label')
# ax.set_ylabel('y label')
plt.plot(x,y)
plt.show() # to remove that line comes with output ([<matplotlib.lines.Line2D at 0x28ece4996d8>])
```

```python
# change tick marks, add legend
x = [0,1,2,3,4]
y = [0,2,4,6,8]
# fig, ax = plt.subplots()
plt.title('Sample Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20,'color':'blue'})
plt.xlabel('X axis')
plt.ylabel('Y axis')
# ax.set_title("Sample Graph")
# ax.set_xlabel('x label')
# ax.set_ylabel('y label')

plt.xticks([0,1,2,3])  # change tick marks
plt.yticks([0,2,4,6,8,10])

# plt.plot(x,y, label='2x',color='red',linewidth=3,marker='.',markersize=10,markeredgecolor='blue',linestyle="--") # label name
# use shorthand notation for this
# [color,marker,line]
plt.plot(x,y, 'r.-',label='2x')

plt.legend() # add legend

plt.show() # to remove that line comes with output ([<matplotlib.lines.Line2D at 0x28ece4996d8>])
```

```python
# add a second line to the graph
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# resize the graph
plt.figure(figsize=(5,3),dpi=300) # dpi pixels per inch, recommended 300

plt.title('Sample Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20,'color':'blue'})
plt.xlabel('X axis')
plt.ylabel('Y axis')

x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,label='x^2')
# print(x2)

plt.xticks([0,1,2,3])  
plt.yticks([0,2,4,6,8,10])

plt.plot(x,y, 'r.-',label='2x')

plt.legend()
plt.savefig('my_graph.png',dpi=300) # save the graph
plt.show() 
```
